```python
from src.main import *
import numpy as np
import matplotlib.pyplot as plt


X, y = generate_random_linear_dataset(num_samples=150, num_features=2, noise_std_dev=3.0, random_seed=42)
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)

initial_weights = np.zeros(X.shape[1] + 1)
trained_weights, final_cost = compute_linear_regression(X, y, alpha=0.01, max_iterations=2000)
print("Trained Weights:", trained_weights)
print("Final Cost:", final_cost)

cost = compute_cost(np.hstack([np.ones((X.shape[0], 1)), X]), y, trained_weights)
print("Cost on generated data:", cost)

compute_r_squared(X, y, trained_weights)


plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], y, label='Generated Data')
# For a simple 2-feature case
if X.shape[1] == 2:
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x_line = np.linspace(x_min, x_max, 100)
    predicted_y = trained_weights[0] + trained_weights[1] * x_line + trained_weights[2] * 0
    plt.plot(x_line, predicted_y, color='red', label='Linear Regression')

plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.title('Linear Regression on Generated Data')
plt.legend()
plt.grid(True)
plt.show()
```
